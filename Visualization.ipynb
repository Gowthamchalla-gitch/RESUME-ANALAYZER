Matplotlib is a Python library used for data visualization — that means it helps you turn your data into graphs, plots, and charts.

Seaborn is another Python data visualization library, built on top of Matplotlib — but it’s easier to use and makes beautiful charts with better style and color automatically.

```python
import matplotlib.pyplot as plt
#pyplot is a module inside Matplotlib
import seaborn as sns
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Study_Based_Recommendation/student_habits_performance.csv')
```

```python
# Histogram: Distribution of study hours
plt.figure(figsize=(8,5))
plt.hist(df['study_hours_per_day'], bins=20, edgecolor='black')
plt.title('Distribution of Study Hours per Day')
plt.xlabel('Study Hours per Day')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 2. Bar Chart: Average Exam Score by Part-Time Job
plt.figure(figsize=(9,5))
df.groupby('part_time_job')['exam_score'].mean().plot(kind='bar', color=['yellow','green'])
plt.title('Average Exam Score by Part-Time Job')
plt.xlabel('Part-Time Job')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
# Bar Chart: Average Exam Score by Part-Time Job
plt.figure(figsize=(9,5))
df.groupby('part_time_job')['exam_score'].mean().plot(kind='barh', color=['yellow','green'])
plt.title('Average Exam Score by Part-Time Job')
plt.xlabel('Part-Time Job')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
# hexbin (hexagonal binning) plot Chart: Average Exam Score by Part-Time Job
plt.figure(figsize=(9,5))
df.plot(kind='hexbin', x='study_hours_per_day', y='exam_score', gridsize=505, cmap='magma')
plt.title('Average Exam Score by Part-Time Job')
plt.xlabel('Part-Time Job')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
#  Pie Chart: Gender Distribution
plt.figure(figsize=(6,6))
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=0, colors=['purple','yellow','pink'])
#autopct='%1.2f%%' - Displays percentage values on each slice
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Create a donut chart for gender distribution
plt.figure(figsize=(6,6))
values = df['gender'].value_counts()
labels = values.index

# Pie chart with a hole in the center
plt.pie( values, labels=labels, autopct='%1.1f%%',startangle=90, colors=['purple','yellow','pink'],
    wedgeprops={'width':0.4, 'edgecolor':'red'}  # makes the hole
)
plt.title('Gender Distribution (Donut Chart)')
plt.show()
```

```python

# Box Plot: Exam Score by Diet Quality
plt.figure(figsize=(8,5))
sns.boxplot(x='diet_quality', y='exam_score', data=df)
plt.title('Exam Score by Diet Quality')
plt.xlabel('Diet Quality')
plt.ylabel('Exam Score')
plt.show()
```

```python
#  Scatter Plot: Study Hours vs Exam Score
# scatter plots chart used for understanding relationships between variables.
plt.figure(figsize=(8,5))
plt.scatter(df['study_hours_per_day'], df['exam_score'], alpha=0.8)
#alpha=0.6 - Transparency — makes overlapping points visible (0 = fully transparent, 1 = solid).
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")

#  Line Plot: Sleep Hours vs Exam Score (averaged)
plt.figure(figsize=(10,5))
sleep_avg = df.groupby('sleep_hours')['exam_score'].mean().sort_index()
plt.plot(sleep_avg.index, sleep_avg.values, marker='+')
plt.title('Average Exam Score vs Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
# 7. Heatmap: Correlation between numeric variables
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes('number').corr(), annot=True, cmap='coolwarm', fmt=".2f")
#annot=True - Displays the actual correlation values inside the cells.
# fmt=".2f" - Formats numbers to 2 decimal places (e.g., 0.83, -0.45).
#cmap='coolwarm' - Sets the color scheme (red for positive correlation, blue for negative).
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```

```python


# Count Plot: Extracurricular Participation
plt.figure(figsize=(7,5))
sns.countplot(x='extracurricular_participation', data=df, palette='pastel')
plt.title('Count of Students by Extracurricular Participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Count')
plt.show()
```

```python
total_students = len(df)

# Create a blank canvas
fig, ax = plt.subplots(figsize=(6,4))
ax.axis('off')

# Add cards (text boxes)
ax.text(0.05, 0.7, f" Total Students\n{total_students}", fontsize=18, bbox=dict(facecolor='lightblue', alpha=0.8))
#f" Total Students\n{total_students}"- Text displayed (supports new line)

plt.title("Student Dashboard Summary", fontsize=16, pad=20)
plt.show()
```
